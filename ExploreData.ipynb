## Explore Different Information Fields
**Any line beginnig with a letter n the range A-Z or a-z and immediately followed by a colon(:) is an information field**

[For More Information](http://abcnotation.com/wiki/abc:standard:v2.1#information_fields)

```python
fields = dict()
```

```python
for line in lines:
    if len(line) > 0:
        if line[0] in string.ascii_letters and line[1] == ":":
            value = line.split(":")[1:]
            if line[0] in fields.keys():
                if value not in fields[line[0]]:
                    fields[line[0]].append(value)
            else:
                fields[line[0]] = list(value)
```
